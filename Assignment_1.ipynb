```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the dataset to python environment.
data = pd.read_csv(r'C:\Users\Anu\Downloads\Assignment 1\iris.csv')
data
```

```python
# Display the columns in the dataset.
data.columns
```

```python
# Calculate the mean of each column of the dataset.
from pandas.api.types import is_numeric_dtype
for column in data.columns:
    if is_numeric_dtype(data[column]):
        print(' Mean of', column,' = ', data[column].mean())
```

```python
# Check for the null values present in the dataset.
data.isnull().sum()
```

```python
# Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

# Visualization 1
sns.countplot(x='PW',data=data, hue='Classification');
```

```python
# Visualization 2

plt.figure(figsize=(6,4))
plt.scatter(data['SW'],data['PW'],s=10)
plt.title('SW vs PW')
plt.xlabel('SW')
plt.ylabel('PW')
```

```python
# Visualization 3

sns.heatmap(data.corr(),annot=True)
plt.show()
```
